## Object-oriented style 
Use in bigger projects if the code might be reused.

```python
# create 100 points from 0 to 2
x = np.linspace(0,2,100)

# plot data and set the labels for each line
fig, ax = plt.subplots()
ax.plot(x, x, label = 'linear')
ax.plot(x, x**2, label = 'quadratic')
ax.plot(x, x**3, label = 'cubic')

# set x and y
ax.set_xlabel('x')
ax.set_ylabel('f(x)')

# set the title 
ax.set_title('Simple plot at 2:31 at the airport')

# add legend 
ax.legend()
```

## Pyplot style 
For fast plotting in notebooks and interactive projects. 

```python
plt.plot(x, x, label='linear')
plt.plot(x, x**2, label='quadratic')
plt.plot(x, x**3, label='cubic')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.title('Another simple plot at 2:40')
plt.legend()
```

## Recommended plotting function

```python
def my_plotter(ax, data1, data2, param_dict):
    """
    A helper function to make a graph

    Parameters
    ----------
    ax : Axes
        The axes to draw to

    data1 : array
       The x data

    data2 : array
       The y data

    param_dict : dict
       Dictionary of kwargs to pass to ax.plot

    Returns
    -------
    out : list
        list of artists added
    """
    out = ax.plot(data1, data2, **param_dict)
    return out
```

```python
data1, data2, data3, data4 = np.random.randn(4, 100)
fig, ax = plt.subplots(1,1)
my_plotter(ax, data1, data2, param_dict = {'marker' : 'x'})
```

```python
fig, (ax1, ax2) = plt.subplots(1,2)
my_plotter(ax1, data1, data2, param_dict = {'marker' : 'o'})
my_plotter(ax2, data3, data4, param_dict = {'marker' : 'x'})
```
